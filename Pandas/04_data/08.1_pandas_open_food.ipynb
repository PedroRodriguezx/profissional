```python
import pandas as pd
```

```python
import requests
```

```python
from datetime import datetime as dt
```

```python
url = "https://world.openfoodfacts.net/api/v2/product/"
product = '7891000369371'
```

```python
url_completa = url + product
```

```python
resposta = requests.get(url_completa)
```

```python
resposta.status_code
```

```python
crunch = resposta.json()
```

```python
crunch 
```

```python
df = pd.json_normalize(crunch.get('product'))
```

```python
df.head()
```

```python
df_cols = df[['code','product_name','image_front_url','last_modified_t']].copy()
```

```python
df_cols.head()
```

```python
df_cols['last_modified_t'] = pd.to_datetime(df_cols['last_modified_t'], unit='s')
```

```python
df_cols
```

```python
df_cols.to_csv('./data/raw/csv/crunch.csv', index=False)
```
