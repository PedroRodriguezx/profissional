```python
import pandas as pd
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('./database/sqlite/northwind.db')
```

```python
query = """
   SELECT name 
     FROM sqlite_master
    WHERE type='table';
   """

df_all_tables = pd.read_sql_query(query, conn)

df_all_tables.head()
```

```python
query_all_ordes = """
   SELECT *
     FROM Orders
   """
```

```python
query_all_ordes
```

```python
df_all_orders = pd.read_sql_query(query_all_ordes, conn)
```

```python
df_all_orders
```

```python
df_all_orders_by_city = (
    df_all_orders
    .groupby('ShipCity')
    .agg(
        count_orders=('OrderID', 'count')
    )
    .sort_values('count_orders', ascending=False)
)
```

```python
df_all_orders_by_city.head()
```

```python
df_all_orders_by_city.to_sql('OrdersByCityCount', conn, if_exists='replace')
```
